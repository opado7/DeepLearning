### GAN 파생 기술
- GAN은 생성자와 판별자가 서로 대결하면서 학습하는 구조이기 때문에 학습이 매우 불안정
- 한쪽으로 치우친 훈련이 발생하면 성능에 문제가 생겨 정상적인 분류가 불가능
- DCGAN은 GAN 학습에 CNN 사용

### DCGAN
- 생성자 네트워크에서는 노이즈를 입력으로 받아 훈련 데이터셋의 이미지와 같은 해상도를 갖는 이미지를 생성해야 하기 때문에 공간을 확장시켜야 업샘플링이 필요
- 판별자 네트워크는 실제 이미지와 생성자가 생성한 이미지 사이에서 어떤 이미지가 진짜인지 판별해야 하기 때문에 각 이미지의 특성을 추출할 수 있는 합성곱 연산 수행
- 스트라이드 합성곱을 사용하면 특성 추출 가능
- 생성자가 수행하는 Deconvolution Layer는 손실된 정보를 복원함
